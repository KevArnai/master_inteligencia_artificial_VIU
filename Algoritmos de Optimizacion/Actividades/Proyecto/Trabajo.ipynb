Respuesta

- ¿Cual es la función objetivo?
  
La función objetivo mide qué tan buena es una asignación de horarios, y en este problema se define como:
La audiencia total generada por los 10 partidos de la jornada.
Cada partido aporta una audiencia que depende de tres factores:
- la audiencia según las categorías de los equipos
- el coeficiente del horario
- la penalización por coincidencias.


```python
#Función objetivo
def funcion_objetivo(asignacion):
    
    audiencia = [1.3, 1.3, 0.47, 1.3, 0.47, 0.75, 0.9, 0.9, 0.75, 0.9]
    coef_horario = {"V20": 0.40,"S12": 0.55,"S16": 0.70,"S18": 0.80,"S20": 1.00,"D12": 0.45,"D16": 0.75,"D18": 0.85,"D20": 1.00,"L20": 0.40}
    penalizacion = {1: 0.00,2: 0.25,3: 0.45,4: 0.60,5: 0.70,6: 0.75,7: 0.78,8: 0.80,9: 0.80,10: 0.80}
    
    #cuantos partidos hay en cada horario
    conteo = {}
    for h in asignacion:
        conteo[h] = conteo.get(h, 0) + 1
        
    audiencia_total = 0

    for i in range(len(asignacion)):
        horario = asignacion[i]
        base = audiencia[i]
        coef = coef_horario[horario]

        # nº de partidos en ese horario
        coinc = conteo[horario]

        # penalización según tabla
        penal = penalizacion[coinc]

        audiencia_total += base * coef * (1 - penal)

    return audiencia_total
```

- ¿Es un problema de maximización o minimización?
  
Es un problema de maximización ya que queremos encontrar la asignación de horarios que produzca la mayor audiencia posible.


Diseña un algoritmo para resolver el problema por fuerza bruta

Respuesta

```python
import itertools

#Horarios disponibles
horarios = ["V20", "S12", "S16", "S18", "S20", "D12", "D16", "D18", "D20", "L20"]

#Algoritmo fuerza bruta
def fuerza_bruta(horarios):
    mejor_solucion = []
    mejor_audiencia = 0
    
    #Generar todas las asignaciones posibles (10 horarios para cada uno de los 10 partidos)
    for asignacion in itertools.product(horarios, repeat=10):
    
        #Restricciones obligatorias
        if "V20" not in asignacion or "L20" not in asignacion:
            continue
    
        # Calcular audiencia
        audiencia = funcion_objetivo(asignacion)
    
        # Actualizar mejor solución
        if audiencia > mejor_audiencia:
            mejor_audiencia = audiencia
            mejor_solucion = asignacion


    print('Solución:',mejor_solucion)
    print('Audiencia:',mejor_audiencia)

    return mejor_audiencia


#Ejecución algoritmo de fuerza bruta
fuerza_bruta(horarios)
```

(*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta

Respuesta

```python
import random

#Horarios disponibles
horarios = ["V20", "S12", "S16", "S18", "S20", "D12", "D16", "D18", "D20", "L20"]
#Número real de partidos
n_partidos = 10

#Algoritmo busqueda aleatoria

def busqueda_aleatoria(iteraciones):
    mejor_solucion = []
    mejor_audiencia = 0
    for i in range(iteraciones):
    
        #Generar una solución aleatoria
        asignacion = [random.choice(horarios) for i in range(n_partidos)]
        
        #Restricciones obligatorias
        if "V20" not in asignacion or "L20" not in asignacion:
            continue
        
        # Evaluar la solución
        audiencia = funcion_objetivo(asignacion)
        
        # Actualizar la mejor solución
        if audiencia > mejor_audiencia:
            mejor_audiencia = audiencia
            mejor_solucion = asignacion
    
    print('Solución:',mejor_solucion)
    
    return mejor_audiencia
    
```

El algoritmo de búsqueda aleatoria mejora drásticamente la complejidad porque solo evalúa un número limitado de soluciones generadas al azar.
Aunque no garantiza encontrar la solución óptima absoluta, permite obtener soluciones de buena calidad en tiempos muy reducidos, lo que supone una mejora enorme respecto a la fuerza bruta.

(*)Calcula la complejidad del algoritmo

Respuesta

La complejidad del algoritmo de búsqueda aleatoria es O(iteraciones*n_partidos) porque en cada iteración se genera una solución aleatoria de tamaño n_partidos, se comprueban las restricciones y se evalúa la función objetivo, todas ellas operaciones lineales respecto al número de partidos.
Por tanto, la búsqueda aleatoria reduce drásticamente el número de soluciones evaluadas y mejora enormemente la eficiencia del algoritmo.
