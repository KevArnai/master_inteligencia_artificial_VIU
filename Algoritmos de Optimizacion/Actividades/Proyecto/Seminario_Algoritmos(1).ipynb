```python
import random
import pandas as pd
from collections import Counter

# -----------------------
# Datos del problema
# -----------------------

matches = [
    ("Celta", "Real Madrid", "B", "A"),
    ("Valencia", "R. Sociedad", "B", "A"),
    ("Mallorca", "Eibar", "C", "C"),
    ("Athletic", "Barcelona", "B", "A"),
    ("Leganés", "Osasuna", "C", "C"),
    ("Villarreal", "Granada", "B", "C"),
    ("Alavés", "Levante", "B", "B"),
    ("Espanyol", "Sevilla", "B", "B"),
    ("Betis", "Valladolid", "B", "C"),
    ("Atlético", "Getafe", "B", "B"),
]

# Slots disponibles (10)
SLOTS = ["V20", "S12", "S16", "S18", "S20", "D12", "D16", "D18", "D20", "L20"]

# Audiencia base (millones) para el mejor horario (S20)
base_audience = {
    ("A","A"): 2.0,
    ("A","B"): 1.3,
    ("A","C"): 1.0,
    ("B","B"): 0.9,
    ("B","C"): 0.75,
    ("C","C"): 0.47
}

def base(cat1, cat2):
    a, b = sorted([cat1, cat2])
    return base_audience[(a,b)]

# Coeficiente por horario (ponderación)
slot_coef = {
    "V20": 0.4,
    "S12": 0.55, "S16": 0.7, "S18": 0.8, "S20": 1.0,
    "D12": 0.45, "D16": 0.75, "D18": 0.85, "D20": 1.0,
    "L20": 0.4
}

# Penalización por coincidencias:
# Si en un slot hay k partidos => coincidencias = k-1 => penalización %
penalty_by_coinc = {
    0: 0.00,
    1: 0.25,
    2: 0.45,
    3: 0.60,
    4: 0.70,
    5: 0.75,
    6: 0.78,
    7: 0.80,
    8: 0.80
}

def coincidence_factor(k_in_slot):
    """Factor multiplicativo por coincidencias en un slot con k partidos."""
    coincidences = max(0, k_in_slot - 1)
    coincidences = min(coincidences, 8)
    return 1.0 - penalty_by_coinc[coincidences]

# -----------------------
# Evaluación de una solución
# -----------------------

def evaluate(solution):
    """
    solution: lista de slots, uno por partido (len=10). Ej: solution[i] = "D16"
    Devuelve: (score_total, dataframe_detalle)
    """
    counts = Counter(solution)

    rows = []
    total = 0.0

    for i, (home, away, c1, c2) in enumerate(matches):
        slot = solution[i]
        b = base(c1, c2)
        w = slot_coef[slot]
        base_times = b * w

        k = counts[slot]
        cf = coincidence_factor(k)
        final = base_times * cf

        rows.append({
            "Partido": f"{home} - {away}",
            "Categorías": f"{c1}-{c2}",
            "Horario": slot,
            "Base(Mill.)": b,
            "Ponderación": w,
            "Base*Pond": round(base_times, 2),
            "Coincidencias_en_slot": k-1,
            "Factor_coinc": cf,
            "Corrección_Coincid": round(final, 2)
        })
        total += final

    df = pd.DataFrame(rows)
    return total, df.sort_values("Horario")

    example_solution = ["V20","S12","S16","S18","S20","D16","D16","D18","D20","L20"]
score, detail = evaluate(example_solution)

print("Score total (millones):", round(score, 2))
detail
```
